Los cuartiles son los tres valores de la variable estadística que dividen a un conjunto de datos ordenados en cuatro partes iguales.
Q1, Q2 y Q3 determinan los valores correspondientes al 25%, al 50% y al 75% de los datos. Q2 coincide con la mediana.

Los outliers son puntos de datos que están lejos del resto de puntos de datos. En otras palabras, son valores inusuales en un conjunto de datos.
Podemos usar el rango intercuartílico (IQR), varios valores de cuartiles y un factor de ajuste para calcular los límites de lo que constituyen valores atípicos menores y mayores.
Menor y mayor denotan lo inusual del valor atípico en relación con la distribución general de valores.
El IQR es el 50% medio del conjunto de datos. Es el rango de valores entre el tercer cuartil y el primer cuartil (Q3 - Q1).
Podemos tomar los valores IQR, Q1 y Q3 para calcular los outliers para nuestro conjunto de datos.


IQR = Q3 - Q1

Lower 1.5*IQR whisker = Q1 - 1.5 * IQR 

Upper 1.5*IQR whisker = Q3 + 1.5 * IQR

### Verificación de estacionariedad y descomposición STL de la serie

```python
plt.figure(figsize=[15,7])
sm.tsa.seasonal_decompose(df_month.Weighted_Price).plot()
print("Dickey–Fuller test: p=%f" % sm.tsa.stattools.adfuller(df_month.Weighted_Price)[1])
plt.show()
```

```python
df_month['Weighted_Price_box'], lmbda = stats.boxcox(df_month.Weighted_Price)
print("Dickey–Fuller test: p=%f" % sm.tsa.stattools.adfuller(df_month.Weighted_Price)[1])
```

```python
df_month['prices_box_diff'] = df_month.Weighted_Price_box - df_month.Weighted_Price_box.shift(12)
print("Dickey–Fuller test: p=%f" % sm.tsa.stattools.adfuller(df_month.prices_box_diff[12:])[1])
```

```python
# Regular differentiation
df_month['prices_box_diff2'] = df_month.prices_box_diff - df_month.prices_box_diff.shift(1)
plt.figure(figsize=(15,7))

# STL-decomposition
sm.tsa.seasonal_decompose(df_month.prices_box_diff2[13:]).plot()   
print("Dickey–Fuller test: p=%f" % sm.tsa.stattools.adfuller(df_month.prices_box_diff2[13:])[1])

plt.show()
```

```python
# Initial approximation of parameters using Autocorrelation and Partial Autocorrelation Plots
plt.figure(figsize=(15,7))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(df_month.prices_box_diff2[13:].values.squeeze(), lags=48, ax=ax)
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(df_month.prices_box_diff2[13:].values.squeeze(), lags=48, ax=ax)
plt.tight_layout()
plt.show()
```

```python
# Initial approximation of parameters
Qs = range(0, 2)
qs = range(0, 3)
Ps = range(0, 3)
ps = range(0, 3)
D=1
d=1
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)

# Model Selection
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')
for param in parameters_list:
    try:
        model=sm.tsa.statespace.SARIMAX(df_month.Weighted_Price_box, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], 12)).fit(disp=-1)
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
```

```python
# Best Models
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
print(best_model.summary())
```

### P4.- ¿Qué es ADF_Test, ACF, PACF, ARIMA y SARIMA? ¿Para qué se usan?

**ADF_Test**: En estadística y econometría , una prueba de Dickey-Fuller aumentada ( ADF ) prueba la hipótesis nula de que una raíz unitaria está presente en una muestra de series de tiempo . La hipótesis alternativa es diferente según la versión de la prueba que se utilice, pero suele ser estacionariedad o tendencia-estacionaria . Es una versión aumentada de la prueba Dickey-Fuller para un conjunto más grande y complicado de modelos de series de tiempo.


**ACF**: Es una función de autocorrelación que nos da valores de autocorrelación de cualquier serie con sus valores rezagados describe qué tan bien se relaciona el valor presente de la serie con sus valores pasados.


**PACF**: En el análisis de series de tiempo , la función de autocorrelación parcial ( PACF ) da la correlación parcial de una serie de tiempo estacionaria con sus propios valores rezagados, regresó los valores de la serie de tiempo en todos los rezagos más cortos. Contrasta con la función de autocorrelación , que no controla otros retrasos.


**ARIMA**: Es un método estadístico utilizado para el pronóstico de series de tiempo. ARIMA es un acrónimo de AutoRegressive Integrated Moving Average. Es una clase de modelo que captura un conjunto de diferentes estructuras temporales estándar en datos de series de tiempo.


**SARIMA**: Una extensión de ARIMA que admite el modelado directo del componente estacional de la serie.

```python
# Inverse Box-Cox Transformation Function
def invboxcox(y,lmbda):
    if lmbda == 0:
        return(np.exp(y))
    else:
        return(np.exp(np.log(lmbda*y+1)/lmbda))
```
