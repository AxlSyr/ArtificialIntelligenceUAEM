## Definiciones Básicas (Introducción):

### 1.- ¿Qué es un compilador y un interprete?

Un compilador es un programa que traduce un programa escrito en un lenguaje de programación de alto nivel que es adecuado para programadores humanos al lenguaje de máquina de bajo nivel que requieren las computadoras. Durante este proceso, el compilador también intentará detectar e informar errores obvios del programador.

Un intérprete es un programa que toma el árbol de sintaxis abstracta de un programa
y lo ejecuta inspeccionando el árbol de sintaxis para ver qué se necesita hacer. Esto es
similar a cómo un humano evalúa una expresión matemática: insertamos los valores
de variables en la expresión y evaluarla poco a poco, comenzando con la más interna
paréntesis y saliendo hasta que tengamos el resultado de la expresión. Luego podemos repetir el proceso con otros valores para las variables.

### 2.- ¿Cuál es la diferencia entre ellos?

Un intérprete es otra forma de implementar un lenguaje de programación. La interpretación comparte muchos aspectos con la compilación. La lectura, el análisis sintáctico y la verificación de tipos se realizan en un intérprete como en un compilador. Pero en lugar de generar código a partir del árbol de sintaxis, el árbol de sintaxis se procesa directamente para evaluar expresiones y
ejecutar sentencias, etc. Es posible que un intérprete deba procesar la misma pieza
del árbol de sintaxis (por ejemplo, el cuerpo de un bucle) muchas veces y, por lo tanto, la interpretación suele ser más lenta que la ejecución de un programa compilado. Pero escribir un intérprete es a menudo más simple que escribir un compilador y el intérprete es más fácil de mover a una máquina diferente, por lo que para aplicaciones donde la velocidad no es esencial, a menudo se utilizan intérpretes.

### 3.- ¿Qué es un lenguaje de programación?

Un lenguaje de programación es un lenguaje formal que comprende un conjunto de instrucciones que producen varios tipos de resultados. Los lenguajes de programación se utilizan en la programación de computadoras para implementar algoritmos.

## Definiciones Básicas (Scanner):

### 1.- ¿Qué es un analizador léxico?

Es la fase del analizador que se encarga de revisar el texto de entrada con base en una gramática dada. Y en caso de que el programa de entrada sea válido, suministra el  ́arbol sintáctico que lo reconoce.

En teoría, se supone que la salida del analizador sintáctico es alguna representación del  ́arbol sintáctico que reconoce la secuencia de tokens
suministrada por el analizador léxico.

### 2.- ¿Cuál es la función de un Scanner?

Un scanner tomará como entrada una cadena de letras individuales y dividirá esta cadena en tokens. Además, filtrará lo que separe los tokens (los llamados espacios en blanco), es decir, los caracteres de diseño (espacios, nuevas líneas, etc.) y los comentarios.

El objetivo principal del análisis léxico es facilitar la vida de la siguiente fase de análisis de sintaxis. Un scanner debe hacer lo siguiente:



1. Encontrar tokens
2. Instalar un nuevo identificador en la tabla de símbolos y devolver la dirección
3. Devolver la dirección de la tabla de símbolos de un identificador previamente instalado
4. Producir una lista numerada del programa fuente.
5. Generar mensajes de error para errores léxicos.





### 3.- ¿Qué es un autómata y qué elementos los componen?

Un autómata finito A = (I, O, S, f, g, σ) es una máquina de estado finito en la que el conjunto de símbolos de salida es {0, 1} y donde el estado actual determina la última salida. Los estados para los que la última salida fue 1 se denominan estados de aceptación.

En un sentido abstracto, es una máquina que tiene un número finito de estados y un número finito de transiciones entre ellos. Una transición entre estados generalmente se etiqueta con un carácter del alfabeto de entrada, pero también usaremos transiciones marcadas con ε, las llamadas transiciones épsilon. Se puede utilizar un autómata finito para decidir si una cadena de entrada es un miembro de algún conjunto particular de cadenas. Para hacer esto, seleccionamos uno de los estados del autómata como estado inicial.


### 4.- ¿Qué es una expresión Regular?

Dado un alfabeto, describiremos conjuntos de cadenas mediante expresiones regulares, una notación algebraica que es compacta y fácil de usar y comprender para los humanos. La idea es que las expresiones regulares que describen conjuntos simples de cadenas se pueden combinar para formar expresiones regulares que describen conjuntos de cadenas más complejos.
